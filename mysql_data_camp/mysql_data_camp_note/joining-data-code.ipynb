```python
# import libraries
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Establishing connection
conn = mysql.connector.connect(
    user='root',
    password='Password1234',
    host='localhost',
    database='classicmodels'
)
cursor = conn.cursor()
```

```python
# view the tables in the database

query = """

SELECT * 
FROM customers

"""

# Execute the query
cursor.execute(query)
result = cursor.fetchall()

# Creating DataFrame for neat presentation
df = pd.DataFrame(result, columns=cursor.column_names)
df
```

```python
# view the tables in the database

query = """


SELECT  
        c.customerNumber as 'customer Number', 
        c.customerName  as 'customer Name', 
        c.contactLastName as 'contact Last Name',
        o.orderNumber as 'order Number'
from customers as c
inner join orders as o
on c.customerNumber = o.customerNumber
where o.orderDate between '2003-01-01' and '2003-12-31'

"""

# Execute the query
cursor.execute(query)
result = cursor.fetchall()

# Creating DataFrame for neat presentation
df = pd.DataFrame(result, columns=cursor.column_names)
df
```

```python
# view the tables in the database

query = """


SELECT  
        c.customerNumber as 'customer Number', 
        c.customerName  as 'customer Name', 
        c.contactLastName as 'contact Last Name',
        o.orderNumber as 'order Number'
from customers as c
inner join orders as o
on c.customerNumber = o.customerNumber
where o.orderDate between '2003-01-01' and '2003-12-31'
union 
select customerNumber, customerName, contactLastName, null
from customers
where customerNumber not in (select customerNumber from orders)

"""

# Execute the query
cursor.execute(query)
result = cursor.fetchall()

# Creating DataFrame for neat presentation
df = pd.DataFrame(result, columns=cursor.column_names)
df
```
